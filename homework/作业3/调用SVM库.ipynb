```python
# 导入需要的库
from sklearn import datasets # 自带数据库
from sklearn.svm import SVC  # 引入svm库里的svc分类工具
from sklearn.model_selection import train_test_split # 引入分类函数
from sklearn.metrics import accuracy_score # 引入判断正确率函数

# 加载系统自带的鸢尾花数据
iris = datasets.load_iris()

X = iris.data  # 四个参数
y = iris.target# 花的种类

# 拆分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)# test_size越大，准确率越低，相当于30%用于测试，70%用于训练模型

# 调用 SVM 库！自动用4个参数分类！
model = SVC()
model.fit(X_train, y_train)  # 训练：自动学习4个特征

# 预测
y_pred = model.predict(X_test)

# 输出结果
print("四个参数分类的准确率：", accuracy_score(y_test, y_pred))
```
